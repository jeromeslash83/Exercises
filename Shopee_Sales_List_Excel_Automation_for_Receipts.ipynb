```python
import os #for working with folders.
import pandas as pd #for loading and working with the data itself.
import warnings
```

```python
path = '/content'
```

```python
excel_file = [file for file in os.listdir(path) if file.endswith('.xlsx')]
dataframes_completed = []
with warnings.catch_warnings(record=True):
    warnings.simplefilter("ignore", UserWarning)
    for file in excel_file:
      file_path = os.path.join(path, file)
      df = pd.read_excel(file_path)
      dataframes_completed.append(df)


# Now since all the DataFrames have been added in the 'dataframes_completed' list, this code will concatenate all the DataFrames into one.
df = pd.concat(dataframes_completed, ignore_index=True)
```

```python
df.columns
```

```python
col_to_remove = ['Order Status', 'Return / Refund Status','Shipping Option', 'Shipment Method',
       'Estimated Ship Out Date', 'Ship Time','Parent SKU Reference No.', 'SKU Reference No.', 'Variation Name', 'Original Price', 'Deal Price',
       'Total Discount(PHP)',
       'Price Discount(from Seller)(PHP)', 'Shopee Rebate(PHP)',
       'SKU Total Weight', 'Number of Items in Order', 'Order Total Weight',
       'Seller Voucher(PHP)', 'Seller Absorbed Coin Cashback',
       'Shopee Voucher(PHP)', 'Bundle Deals Indicator(Y/N)',
       'Shopee Bundle Discount(PHP)', 'Seller Bundle Discount(PHP)',
       'Shopee Coins Offset(PHP)', 'Credit Card Discount Total(PHP)', 'Buyer Paid Shipping Fee',
       'Shipping Rebate Estimate', 'Reverse Shipping Fee', 'Service Fee',
       'Grand Total', 'Estimated Shipping Fee','Remark from buyer',
       'Order Complete Time', 'Note']

df.drop(columns=col_to_remove, inplace=True)
```

```python
df.shape
```

```python
agg_funcs = {
    'Order Creation Date': 'first',  # Keep the first date in the group
    'Order Paid Time': 'first',  # Keep the first time in the group
    'Product Name': ', '.join,  # Join product names with ', '
    'Quantity': ', '.join,  # Join quantities with ', '
    'Product Subtotal': ', '.join,  # Join subtotals with ', '
    'Username (Buyer)': 'first',  # Keep the first username in the group
    'Receiver Name': 'first',  # Keep the first receiver name in the group
    'Phone Number': 'first',  # Keep the first phone number in the group
    'Delivery Address': 'first',  # Keep the first delivery address in the group
    'Town': 'first',  # Keep the first town in the group
    'District': 'first',  # Keep the first district in the group
    'Province': 'first',  # Keep the first province in the group
    'Region': 'first',  # Keep the first region in the group
    'Country': 'first',  # Keep the first country in the group
    'Zip Code': 'first'  # Keep the first zip code in the group
}

df['Quantity']=df['Quantity'].astype('str')
df['Product Subtotal']=df['Product Subtotal'].astype('str')
```

```python
df_grouped = df.groupby(['Order ID', 'Tracking Number*']).agg(agg_funcs).reset_index()
```
